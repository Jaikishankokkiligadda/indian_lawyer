```python
# Retriever And Chain 
from langchain_community.document_loaders import TextLoader

loader = TextLoader("court_order.txt")
text_documents = loader.load()
text_documents 
```

```python
# create chukenks 
from langchain.community.vectorstores import Chroma
db = Chroma.from_documents(text_documents, collection_name="court_orders")
retriever = db.as_retriever()
retriever.get_relevant_documents("What is the court order about?")

```
